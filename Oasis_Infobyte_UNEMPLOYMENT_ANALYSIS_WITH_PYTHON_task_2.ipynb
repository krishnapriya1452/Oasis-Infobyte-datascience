```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/content/Unemployment in India.csv")
data
```

```python
data = data.dropna()
```

```python
data.info
```

```python
data.describe()
```

```python
data[' Date'] = pd.to_datetime(data[' Date'])
```

```python
plt.figure(figsize=(15,5))
plt.bar(data[' Date'], data[' Estimated Unemployment Rate (%)'])
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate in India')
plt.show()
```

```python
#displaying the advanced visualization between the unemployment rate and on particular date
plt.figure(figsize=(15,5))
sns.lineplot(x=' Date',y = ' Estimated Unemployment Rate (%)',data=data)
```

```python
#caluclating the mean and median of unemployment rate
mean_unemployment = data[' Estimated Unemployment Rate (%)'].mean()
median_unemployment = data[' Estimated Unemployment Rate (%)'].median()
print('Mean Unemployment Rate: ',{mean_unemployment})
print('Median Unemployment Rate:' ,{median_unemployment})
```
